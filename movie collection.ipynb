```python
X=movie.loc[:,movie.columns!="Start_Tech_Oscar"]

y=movie['Start_Tech_Oscar']

from sklearn.linear_model import LogisticRegression

clf_lr=LogisticRegression()

clf_lr.fit(X,y)

clf_lr.coef_

clf_lr.intercept_

import statsmodels.api as sn

X_cons=sn.add_constant(X)

X_cons.head()

import statsmodels.discrete.discrete_model as sm #in tis beta zero is by default 0.

logistic_m=sm.Logit(y,X_cons).fit()

logistic_m.summary()

# predicting and confusion matrix

clf_lr.predict_proba(X)  #the first row is probability that the house not soldand te second row is probability that the house get sold.

y_pred=clf_lr.predict(X)
y_pred

from sklearn.metrics import confusion_matrix

confusion=confusion_matrix(y,y_pred) #a11(pred =0,actual=0(true negatives)), a12(pred=1,actual=0(false positive)),a21(pred=0,actual=1(false negative)),a22(pred=1,pred=1(true positive))
print(confusion)
y_pred_02=(clf_lr.predict_proba(X)[:,1]>=[0.2]) #if we accept if sold prob is 0.2
print(y_pred_02)

confusion_matrix1=confusion_matrix(y,y_pred_02)
print(confusion_matrix1)
# performance matrix
## precision =true positive/(true positive + false positive)
## recall=true positive/(true positive + false negative)
## AUC(ROC)

from sklearn.metrics import precision_score,recall_score,roc_auc_score,accuracy_score

print(precision_score(y,y_pred))

print(recall_score(y,y_pred))

print(roc_auc_score(y,y_pred))

true_positives1 = int(confusion_matrix1[:1,:1])
true_negatives1 = int(confusion_matrix1[-1:, -1:])
false_positives1 = int(confusion_matrix1[:1,-1:])
false_negatives1 = int(confusion_matrix1[-1:, :-1])

# Sensitivity
sensitivity1 = round(true_positives1 / (true_positives1 + false_negatives1), 2)
print(sensitivity1)

# Specificity
specificity1 = round(true_negatives1 / (true_negatives1 + false_positives1), 2)
print(specificity1)

from sklearn.metrics import accuracy_score
print(round(accuracy_score(y_pred, y), 2))
```
